#### EDA - All csv files

```python
base_path = "abfss://f1-data@prstrgacc.dfs.core.windows.net/bronze/"
```

```python
from pyspark.sql.functions import col, count, when

def perform_eda(df, df_name):
        
    print("\nSchema:")
    df.printSchema()
    
    print("\nSample Data:")
    df.show(10, False)
    
    print("\nTotal Record Count:")
    print(df.count())
    
    print("\nNull Value Count:")
    df.select([
        count(when(col(c).isNull(), c)).alias(c)
        for c in df.columns
    ]).show()
    
    print("\nDistinct Values per Column:")
    for c in df.columns:
        print(f"{c}:", df.select(c).distinct().count())

    
```

Read All CSV Files

```python
circuits_df = spark.read.option("header",True).option("inferSchema",True).csv(base_path + "circuits.csv")

races_df = spark.read.option("header",True).option("inferSchema",True).csv(base_path + "races.csv")

drivers_df = spark.read.option("header",True).option("inferSchema",True).csv(base_path + "drivers.csv")

constructors_df = spark.read.option("header",True).option("inferSchema",True).csv(base_path + "constructors.csv")

results_df = spark.read.option("header",True).option("inferSchema",True).csv(base_path + "results.csv")
```

Run EDA for Each Dataset

```python
perform_eda(circuits_df, "Circuits")
```

```python
perform_eda(races_df, "Races")
```

```python
perform_eda(drivers_df, "Drivers")
```

```python
perform_eda(constructors_df, "Constructors")
```

```python
perform_eda(results_df, "Results")
```
