## Race Results Analytics & Insights

```python
race_results_df = spark.table("f1_catalog.f1_gold.race_results")
```

```python
race_results_df.show()
```

Top Drivers by Total Points

```python
driver_points_df = race_results_df.groupBy("driver_name") \
.sum("points") \
.withColumnRenamed("sum(points)", "total_points") \
.orderBy("total_points", ascending=False)
```

```python
display(driver_points_df)
```

Drivers with Most Wins : Shows drivers with the most race wins.

```python
driver_wins_df = race_results_df.filter("position = 1") \
.groupBy("driver_name") \
.count() \
.withColumnRenamed("count", "wins") \
.orderBy("wins", ascending=False)

display(driver_wins_df)
```

Total Points by Constructor: Shows best performing teams.

```python
constructor_points_df = race_results_df.groupBy("constructor_name") \
.sum("points") \
.withColumnRenamed("sum(points)", "total_points") \
.orderBy("total_points", ascending=False)

display(constructor_points_df)
```

Circuits Hosting Most Races:-  ShowsCircuits Hosting Most Races

```python
circuit_races_df = race_results_df.groupBy("circuit_location") \
.count() \
.withColumnRenamed("count", "total_races") \
.orderBy("total_races", ascending=False)

display(circuit_races_df)
```

Race Winners:-  Shows who won each race.

```python
race_winners_df = race_results_df.filter("position = 1") \
.select(
    "race_year",
    "race_name",
    "driver_name",
    "constructor_name"
)

display(race_winners_df)
```
